# 1. User Login Check
### Problem Statement: Given a username and password, check whether login is successful.
username = "admin" \
password = "1234"  \
Requirements:  \
Print "Login Successful" if both username and password match  \
Otherwise print "Invalid Credentials"  \
Real-World Application: Authentication systems



```python
username = "admin"
password = "1234"

#User input
input_username = input("Enter username: ")
input_pass = input("Enter password: ")

#Check
if input_username == username and input_pass == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

# 2. Pass / Fail Analyzer
### Problem Statement: Analyze student results.
marks = [45, 78, 90, 33, 60]

Requirements:   \
A student passes if marks â‰¥ 50   \
Count the total number of pass students   \
Count the total number of fail students   \
Print the final result clearly


```python
#Given
marks = [45,78,90,33,60]
count_pass = 0
count_fail = 0

#check
for i in marks:
    if i >=50:
        count_pass+=1
    else:
        count_fail+=1

print("count of total number of pass students: ",count_pass)
print("count of total number of fail students: ",count_fail)
```

# 3. Simple Data Cleaner
### Problem Statement: Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]

Requirements:  \
Remove extra spaces  \
Convert all names to lowercase  \
Print the cleaned list  \
Real-World Application: Data preprocessing before analysis


```python
#given
names = ["Alice","bob","CHARLIE"]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print(cleaned_names)
```

# 4. Message Length Analyzer
### Problem Statement: Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]

Requirements:   \
Print the length of each message   \
Flag messages longer than 10 characters \
Real-World Application: Text filtering and validation systems


```python
#given
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print("message: ",messages)
    print("length: ",length)
    if length > 10:
        print("messages longer than 10 characters")
print()
```

# 5. Error Message Detector
### Problem Statement: Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements: \
Count the number of "ERROR" entries \
Print the total error count \
Real-World Application: Monitoring and log analysis systems


```python
#given
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count = 0
for log in logs:
    if log == "ERROR":
        count+=1
    
```
